```python
### Merge the datasets on the common key or Id.
### Code_module mean code for which module of course.
### cde_presentation is the code for which module of courseis presented.

```

```python
### Create unique identification for each student by combining the code_module and code_presentation columns.   
student_data['code_module_presentation'] = student_data[['code_module','code_presentation']].apply(lambda x: '_'.join(x), axis=1)   
```

```python
###Create unique identification for each student by combining the id_student and code_module_presentation columns. 
student_data=student_data.astype({'id_student': str})
student_data['code_module_student_presentation'] = student_data[['id_student','code_module_presentation']].apply(lambda x: '_'.join(x), axis=1)
```

```python
uniquevalues = student_data['imd_band'].unique()
print("Unique values in the 'imd_band' column:", uniquevalues)
```

```python
### Replace the unique values with '10-20' with '10-20%'
student_data['imd_band'] = student_data['imd_band'].replace('10-20', '10-20%')
uniquevalues = student_data['imd_band'].unique()
print("Unique values in the 'imd_band' column after replacement:", uniquevalues)
```
